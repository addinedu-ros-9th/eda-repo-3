```python
import pandas as pd
df_raw = pd.read_excel("../data/서울시_시내버스_정류소_구별.xlsx",usecols='B:D')

# 'ID' 열에서 하이픈 제거 후 정수형으로 변환
df_raw['ID'] = df_raw['ID'].str.replace('-', '').astype(int)

# 결과 확인
df_raw
```

```python
print(len(df_raw['자치구'].unique()),df_raw['자치구'].unique())
```

```python
df_local_row = pd.read_csv("../data/버스노선별_정류장별_시간대별_승하차_인원_정보.csv")

df_local_row
```

```python
len(df_local_row['노선번호'].unique())
```

```python
# 예: '버스정류장ARS번호' 열에서 '~' 제거하고 int로 변환
df_local_row['버스정류장ARS번호'] = pd.to_numeric(df_local_row['버스정류장ARS번호'], errors='coerce')

df_local_row = df_local_row.dropna(subset=['버스정류장ARS번호'])
df_local_row['버스정류장ARS번호'] = df_local_row['버스정류장ARS번호'].astype(int)

df_local_row['버스정류장ARS번호'].dtype
```

```python
df_local = df_local_row.iloc[:, 2:-1]
df_local
```

```python
# 두 데이터 연결 (ID ↔ 버스정류장ARS번호)
df_merged = pd.merge(df_raw, df_local, left_on='ID', right_on='버스정류장ARS번호', how='inner')

# 4번째(인덱스 3), 5번째(인덱스 4) 열 이름 가져오기
cols_to_drop = df_merged.columns[[5, 6, 7]]

# 열 제거
df_merged = df_merged.drop(columns=cols_to_drop)

# 결과 확인
df_merged
```

```python
df_merged.info()
```

```python
print(len(df_merged['자치구'].unique()),df_merged['자치구'].unique())
```

```python

# 시간대 리스트 (0~23시), 컬럼명이 실제 데이터에 맞게 구성되어야 함
hours = list(range(24))

# 새로운 통합 컬럼 생성
for hour in hours:
    # 실제 컬럼명 처리 (0시는 '00시', 나머지는 '1시', '2시' 등으로 가정)
    if hour == 0:
        승차_col = "00시승차총승객수"
        하차_col = "00시하차총승객수"
        col_name = "0시"
    else:
        승차_col = f"{hour}시승차총승객수"
        하차_col = f"{hour}시하차총승객수"
        col_name = f"{hour}시"

    df_merged[col_name] = df_merged[승차_col] + df_merged[하차_col]

# 원래 승차/하차 컬럼 삭제 (선택사항)
drop_cols = ["00시승차총승객수", "00시하차총승객수"] + \
            [f"{i}시승차총승객수" for i in range(1, 24)] + \
            [f"{i}시하차총승객수" for i in range(1, 24)]

df_day = df_merged.drop(columns=drop_cols)
```

```python
# 1. 필요한 열만 남기기 (자치구 + 시간대 열들)
columns_to_keep = ['자치구', 'ID'] + [col for col in df_day.columns if col.endswith('시')]
df_cleaned = df_day[columns_to_keep]

# 2. 자치구별 정류소 개수 (ID의 고유값 개수)
stop_counts = df_cleaned.groupby('자치구')['ID'].nunique().reset_index(name='정류소 수')

# 3. 자치구별 시간대별 이용자 수 합산
usage_sums = df_cleaned.drop(columns='ID').groupby('자치구').sum(numeric_only=True).reset_index()

# 4. 정류소 수와 시간대별 이용자 수 합계 합치기
df_result = pd.merge(stop_counts, usage_sums, on='자치구')

# Identify columns with hour data
hour_columns = [col for col in df_result.columns if "시" in col]

# Divide the hour values by 30 and cast to int
df_result[hour_columns] = df_result[hour_columns].div(30).astype(int)
df_result.head()
```

```python
# 우선 심야버스만 필터링
night_bus_df = df_day[df_day['교통수단타입명'] == '서울심야버스']

# 중복 제거 (노선번호, 노선명, 자치구 단위로)
unique_night_routes = night_bus_df[['노선번호', '노선명', '자치구']].drop_duplicates()

# 자치구별 심야버스 노선 수 세기
gu_night_bus_count = unique_night_routes.groupby('자치구').size().reset_index(name='심야버스노선수')

# 자치구 기준으로 merge
df_gu_with_night_count = df_result.merge(gu_night_bus_count, on='자치구', how='left')

# 심야버스노선수가 없는 자치구는 0으로 채우기
df_gu_with_night_count['심야버스노선수'] = df_gu_with_night_count['심야버스노선수'].fillna(0).astype(int)
df_gu_with_night_count.head()
```

```python
# 현재 컬럼 목록
cols = df_gu_with_night_count.columns.tolist()

# '심야버스노선수'를 3번째 자리에 삽입
cols.insert(2, cols.pop(cols.index('심야버스노선수')))

# 컬럼 순서 재정렬
df_gu_with_night_count = df_gu_with_night_count[cols]
df_gu_with_night_count
```

```python
import numpy as np

def convert_types(row):
    new_row = []
    for item in row:
        if isinstance(item, (np.int64, np.integer)):
            new_row.append(int(item))
        elif isinstance(item, (np.float64, np.floating)):
            new_row.append(float(item))
        else:
            new_row.append(item)
    return tuple(new_row)

data = [convert_types(row) for row in df_db.values]
```
