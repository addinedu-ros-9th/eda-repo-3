```python
import pandas as pd
import re
```

```python
seoul_metro_station = pd.read_csv("../data/서울교통공사_자치구별 지하철역 정보_20241127.CSV", encoding="euc-kr")
seoul_metro_station = seoul_metro_station.copy()
```

```python
# 데이터 로드
seoul_metro_station = pd.read_csv("../data/서울교통공사_자치구별 지하철역 정보_20241127.CSV", encoding="euc-kr")

# '해당역(호선)' 칼럼에서 괄호 안의 숫자만 추출하여 '호선' 칼럼 생성
seoul_metro_station['호선'] = seoul_metro_station['해당역(호선)'].apply(lambda x: re.findall(r'\((\d+)\)', str(x)))

# 중복을 제거하고, 리스트로 저장
seoul_metro_station['호선'] = seoul_metro_station['호선'].apply(lambda x: list(set(x)))

# 결과 확인
seoul_metro_station
```

## 코레일 주소데이터

```python
korail_station = pd.read_csv("../data/92. 코레일 주소데이터.csv", encoding="euc-kr")
korail_station = korail_station.copy()
```

```python
# 서울 데이터만 
korail_station = korail_station[korail_station['지번주소'].str.contains('서울', na=False)]


# "숫자+호선" 패턴이 포함된 행을 제외
korail_station = korail_station[~korail_station['선명'].str.contains(r'\d+호선', regex=True)]


korail_station
```

### 코레일 자치구별 정리

```python
# '자치구' 컬럼 추가 (구 추출)
korail_station['자치구'] = korail_station['도로명주소'].str.extract(r'([가-힣]+구)').fillna('')

# 그룹화하여 새로운 데이터프레임 생성
korail_summary = korail_station.groupby('자치구').agg(
    개수=('자치구', 'count'),  # 해당 구에 속하는 행 개수
    선명=('선명', lambda x: list(set(x)))  # 중복 제거된 선명 리스트
).reset_index()

# 출력
seoul_metro_station
```

## 서울교통공사, 코레일 데이터 병합

```python

# '자치구'를 기준으로 두 데이터프레임 병합
merged_df = pd.merge(seoul_metro_station, korail_summary, on='자치구', how='inner')

# 역 개수 합산
merged_df['총 역 개수'] = merged_df['역개수'] + merged_df['개수']

# 호선 정보 병합 (리스트끼리 합치기)
merged_df['호선 정보'] = merged_df.apply(lambda row: row['호선'] + row['선명'], axis=1)

# 필요한 컬럼만 선택
final_df = merged_df[['자치구', '총 역 개수', '호선 정보']]

# 출력
final_df
```

```python
# '자치구'가 '구'로 끝나는 데이터만 필터링
seoul_metro_station = seoul_metro_station[seoul_metro_station['자치구'].str.endswith('구')]
korail_summary = korail_summary[korail_summary['자치구'].str.endswith('구')]

# '자치구'를 기준으로 outer join 수행 (seoul_metro_station을 기준으로)
merged_df = pd.merge(seoul_metro_station, korail_summary, on='자치구', how='left')

# korail_summary에 없는 자치구는 개수를 0, 선명은 빈 리스트로 처리
merged_df['개수'] = merged_df['개수'].fillna(0).astype(int)
merged_df['선명'] = merged_df['선명'].apply(lambda x: x if isinstance(x, list) else [])

# 역 개수 합산
merged_df['총 역 개수'] = merged_df['역개수'] + merged_df['개수']

# 호선 정보 병합 (리스트 합치기)
merged_df['호선 정보'] = merged_df.apply(lambda row: row['호선'] + row['선명'], axis=1)

# 필요한 컬럼만 선택
final_df = merged_df[['자치구', '총 역 개수', '호선 정보']]

# 출력
final_df
```

```python
# '호선 정보' 칼럼의 항목 수 계산하여 새로운 컬럼 '호선_수'에 추가
final_df['호선_수'] = final_df['호선 정보'].apply(len)

final_df
```

```python
# '호선 정보' 칼럼의 리스트 데이터를 숫자 순서대로 정렬
final_df['호선 정보'] = final_df['호선 정보'].apply(lambda x: sorted(x, key=lambda y: (int(y) if y.isdigit() else float('inf'))))

final_df
```

```python
import numpy as np

def convert_types(row):
    new_row = []
    for item in row:
        if isinstance(item, (np.int64, np.integer)):
            new_row.append(int(item))
        elif isinstance(item, (np.float64, np.floating)):
            new_row.append(float(item))
        else:
            new_row.append(item)
    return tuple(new_row)

data = [convert_types(row) for row in df_db.values]

cursor.executemany(sql, data)
conn.commit()
```
