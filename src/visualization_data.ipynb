# 점수 다시


```python

def normalize_score(series, reverse=False):
    max_val = series.max()
    min_val = series.min()
    if reverse:
        return 1 + 9 * (max_val - series) / (max_val - min_val)
    else:
        return 1 + 9 * (series - min_val) / (max_val - min_val)

# 점수 계산
df_scores = pd.DataFrame()
df_scores['gu_id'] = ratio_df['gu_id']
df_scores['df_aged_population'] = normalize_score(ratio_df['df_aged_population'])        # 높을수록 좋음
df_scores['df_gu_disabled'] = normalize_score(ratio_df['df_gu_disabled'])                # 높을수록 좋음
df_scores['df_gu_car'] = normalize_score(ratio_df['df_gu_car'], reverse=True)            # 낮을수록 좋음
df_scores['df_gu_bus_station'] = normalize_score(ratio_df['df_gu_bus_station'], reverse=True)
df_scores['df_sstation_count'] = normalize_score(ratio_df['df_sstation_count'], reverse=True)
df_scores['df_gu_emergency'] = normalize_score(ratio_df['df_gu_emergency'], reverse=True)

df_scores
```

```python
gu_list = ['강남구', '강동구', '강북구', '강서구', '관악구', 
           '광진구', '구로구', '금천구', '노원구', '도봉구', 
           '동대문구', '동작구', '마포구', '서대문구', '서초구', 
           '성동구', '성북구', '송파구', '양천구', '영등포구', 
           '용산구', '은평구', '종로구', '중구', '중랑구']


gu_dict = dict(enumerate(gu_list))
df_scores['gu_id'] = df_scores['gu_id'].map(gu_dict)
df_scores
```

## 점수 재산정

```python
def normalize_score(series, reverse=False):
    max_val = series.max()
    min_val = series.min()
    if reverse:
        return 1 + 9 * (max_val - series) / (max_val - min_val)
    else:
        return 1 + 9 * (series - min_val) / (max_val - min_val)

# 점수 계산
df_scores = pd.DataFrame()
df_scores['gu_id'] = ratio_df['gu_id']
df_scores['df_accident'] = normalize_score(ratio_df['df_accident'])        # 높을수록 좋음
df_scores['df_bicycle'] = normalize_score(ratio_df['df_bicycle'])                # 높을수록 좋음
df_scores['df_night_bus'] = normalize_score(ratio_df['df_night_bus'])                # 높을수록 좋음
df_scores['df_moving_population'] = normalize_score(ratio_df['df_moving_population'])                # 높을수록 좋음
df_scores['df_gu_car'] = normalize_score(ratio_df['df_gu_car'], reverse=True)            # 낮을수록 좋음

df_scores
```

```python
# GeoJSON 데이터를 통해 각 구의 중심을 계산
def get_centroid(geo_data):
    """
    주어진 GeoJSON 데이터에서 폴리곤의 중심 좌표를 구하는 함수.
    """
    geom = shape(geo_data['geometry'])
    centroid = geom.centroid
    return centroid.x, centroid.y
```

```python
# 대상 구 및 조건
target_districts = ['동대문구', '동작구', "서대문구"]
rank_condition = df_scores_weak[df_scores_weak['rank'] <= 3].sort_values(by='rank')['gu_id'].tolist()
targets = target_districts + rank_condition

# 순서대로 정렬된 구 데이터
target_data = df_scores_weak[df_scores_weak['gu_id'].isin(targets)].sort_values(by=['gu_id'], key=lambda x: x.map(lambda y: target_districts.index(y) if y in target_districts else rank_condition.index(y) + len(target_districts))).iloc[:, 1:-2]

# 제목 준비
titles = df_scores_weak[df_scores_weak['gu_id'].isin(targets)].sort_values(by=['gu_id'], key=lambda x: x.map(lambda y: target_districts.index(y) if y in target_districts else rank_condition.index(y) + len(target_districts)))
titles = [f"{title} (Rank {df_scores_weak[df_scores_weak['gu_id'] == title]['rank'].values[0]})" for title in titles['gu_id']]

# 2행 3열로 레이아웃 변경
fig, axes = plt.subplots(2, 3, figsize=(30, 12), subplot_kw=dict(polar=True), gridspec_kw={'wspace': 1.2})

labels = ['노인인구 점수', '장애인구 점수', '자차 점수', '버스 점수', '지하철 점수', '응급실 점수']

#강남구, 마포구는 첫 번째 행에 배치하고 나머지 구들은 두 번째 행에 배치
for i, (df, title) in enumerate(zip(target_data.iterrows(), titles)):
    # if "강남구" in title :
    #     row, col = 0, 0
    # elif "마포구" in title:
    #     row, col = 1, 0
    # else:  # 두 번째 행에는 나머지 구들 배치
    #     if "1" in title or "2" in title:
    #         row, col = 0, i - 1
    #     elif "3" in title or "4" in title:
    #         row, col = 1, i - 3
    if "동대문구" in title :
        row, col = 0, 0
    elif "동작구" in title :
        row, col = 0, 1
    elif "서대문구" in title :
        row, col = 0, 2
    else:
        row, col = 1, i-3

    # 레이더 차트 값 설정
    values = df[1].values.flatten()[:len(labels)]
    angles = [n / float(len(labels)) * 2 * np.pi for n in range(len(labels))]
    angles += angles[:1]
    values = np.append(values, values[0])

    # # 차트 그리기

    # 강남구, 마포구는 빨간색으로 표시
    if "동대문구" in title or "동작구" in title or "서대문구" in title:
        axes[row, col].fill(angles, values, color='red', alpha=0.3)
        axes[row, col].plot(angles, values, color='red', linewidth=2)
    else:
        axes[row, col].fill(angles, values, color='black', alpha=0.3)
        axes[row, col].plot(angles, values, color='black', linewidth=2)

    axes[row, col].set_yticklabels([])
    axes[row, col].set_xticks(angles[:-1])
    axes[row, col].set_xticklabels(labels, fontsize=10)
    axes[row, col].set_title(title, fontsize=14, fontweight="bold", pad=20)

plt.show()
```
