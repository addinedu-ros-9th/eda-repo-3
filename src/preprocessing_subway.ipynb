```python
train_cst_filtered = train_cst_filtered.copy()
# 자료형 일치 (혹시 다를 경우 대비)
train_cst_filtered["행정동_ID"] = train_cst_filtered["행정동_ID"].astype(int)
local_code["행정동_ID"] = local_code["행정동_ID"].astype(int)

# train_cst_filtered에 "구" 컬럼 추가 (merge 이용)
train_cst_filtered = train_cst_filtered.merge(local_code[["행정동_ID", "자치구_명칭"]], 
                                              left_on="행정동_ID", 
                                              right_on="행정동_ID", 
                                              how="left")



# 결과 확인
train_cst_filtered
```

```python
# Nan값 0으로
train_cst_filtered = train_cst_filtered.fillna(0)
```

```python
train_cst_filtered 
```

```python
train_cst_filtered = train_cst_filtered.drop(columns=["기준_날짜"])
```

```python
train_cst_filtered
```

```python
# 칼럼 이름 변경
train_cst_filtered.columns = [col.split('_')[-1] if '지하철_승객_수' in col else col for col in train_cst_filtered.columns]

train_cst_filtered
```

```python
# 자치구 명칭으로 정리
train_cst_filtered["자치구_명칭"] = train_cst_filtered["자치구_명칭"].astype(str)
train_cst_filtered = train_cst_filtered.sort_values(by="자치구_명칭")
train_cst_filtered
```

```python
# 행정동_ID 삭제
train_cst_filtered.drop(columns = ["행정동_ID"], inplace=True)  
```

```python
# 숫자형 칼럼만 선택
numeric_cols = train_cst_filtered.select_dtypes(include=['number']).columns


# 합계 계산
train_cst_filtered_sum = train_cst_filtered.groupby("자치구_명칭", as_index=False)[numeric_cols].sum()

# 평균 계산
# 새로운 데이터프레임에 숫자형 칼럼을 date_difference로 나누어 저장
train_cst_filtered_avg = train_cst_filtered_sum.copy()
train_cst_filtered_avg[numeric_cols] = train_cst_filtered_avg[numeric_cols] / date_difference


```

```python
import numpy as np

def convert_types(row):
    new_row = []
    for item in row:
        if isinstance(item, (np.int64, np.integer)):
            new_row.append(int(item))
        elif isinstance(item, (np.float64, np.floating)):
            new_row.append(float(item))
        else:
            new_row.append(item)
    return tuple(new_row)

data = [convert_types(row) for row in df_db.values]

cursor.executemany(sql, data)
conn.commit()
```
