```python
import pandas as pd
df_raw = pd.read_csv("../data/서울시 장애유형별. 장애정도별 장애인 등록현황.csv")

df_raw
```

```python
# 1. "심하지 않은 장애" 열 제거
df_filtered = df_raw.loc[:, ~df_raw.iloc[0].str.contains("심하지 않은 장애", na=False)]

# 2. 첫 번째 행을 컬럼으로 지정
df_filtered.columns = df_filtered.iloc[0]
df_filtered = df_filtered.drop(index=0).reset_index(drop=True)

# 3. '합계' 행 제거 (필요하다면)
df_filtered = df_filtered[df_filtered.iloc[:, 0] != '합계']

# 4. 구 이름을 인덱스로 설정
df_filtered = df_filtered.set_index(df_filtered.columns[0])

df_filtered.head()
```

```python
# 각 구별 심한 장애 인원 합산
df_filtered["심한 장애 합계"] = df_filtered.iloc[:, 1:].apply(lambda row: row.astype(str).str.replace(",", "").astype(int).sum(), axis=1)

df_summary = df_filtered[["심한 장애 합계"]]

df_summary.head()
```

```python
# 인덱스를 컬럼으로 되돌리기
df_summary = df_summary.reset_index()
df_summary.head()
```

```python
df_summary.columns = ['구', '심한 장애 합계']
df_summary.head()
```

```python
df_summary.to_csv("../data/서울시_구별_장애인.csv")
```

```python
import numpy as np

def convert_types(row):
    new_row = []
    for item in row:
        if isinstance(item, (np.int64, np.integer)):
            new_row.append(int(item))
        elif isinstance(item, (np.float64, np.floating)):
            new_row.append(float(item))
        else:
            new_row.append(item)
    return tuple(new_row)

data = [convert_types(row) for row in df_db.values]

cursor.executemany(sql, data)
conn.commit()
```
